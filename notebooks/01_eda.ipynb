```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

plt.style.use("seaborn-v0_8")
%matplotlib inline
```

```python
data_path = Path("data") / "raw" / "vendor_payments.csv"
df = pd.read_csv(data_path, parse_dates=["payment_date"])
df.head()
```

```python
df.describe(include="all").transpose()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(x=df["amount"], ax=ax)
ax.set_title("Distribution of Vendor Payment Amounts (₹)")
fig.tight_layout()
output_fig = Path("notebooks") / "fig_vendor_amounts.png"
fig.savefig(output_fig)
output_fig
```
